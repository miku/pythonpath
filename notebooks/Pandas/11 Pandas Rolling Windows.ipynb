# 11 Pandas Rolling Windows


```python
import numpy as np
import pandas as pd

%matplotlib inline
```

```python
np.random.seed(0)
```

```python
s = pd.Series(np.random.randn(1000), index=pd.date_range('2000-01-01', periods=1000))
w = s.cumsum()
```

```python
w.plot()
```

Rolling window calculate aggregates over fixed window sizes. The main parameter is the size of the window. Similar to `df.groupby` the aggregation function is chained.

> Size of the moving window. This is the number of observations used for calculating the statistic. Each window will be a fixed size.

```python
w.plot()
w.rolling(50).mean().plot()
```

It is possible to smooth out a curve.

```python
w.plot()
w.rolling(50).max().plot()
```

```python
w.plot()
w.rolling(100).min().plot()
```

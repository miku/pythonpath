```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

```python
np.random.seed(0)
```

```python
data = np.random.uniform(-1, 1, (30, 3))
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.cumsum().plot()
```

```python
sns.pairplot(df)
```

```python
sales = pd.DataFrame(np.random.randint(10, 1000, (10, 10)),
                     index=pd.date_range(start="2000-01-01", periods=10),
                     columns=["Berlin", "London", "Paris", "Tokyo", "Rio",
                              "Praha", "Oslo", "Madrid", "Vienna", "Taipeh"])

sales["Rio"] = np.random.randint(10, 2000, 10)
```

```python
sales
```

```python
sales.cumsum()
```

```python
sns.heatmap(sales.cumsum(), xticklabels=sales.index.format("%Y-%m-%d")[1:])
```

If we transpose the DataFrame, we can move the time to the x-axis.

```python
sns.heatmap(sales.cumsum().T, xticklabels=sales.index.format("%Y-%m-%d")[1:])
```

Many color palettes are shipped with the library.

```python
sns.heatmap(sales.cumsum().T, xticklabels=sales.index.format("%Y-%m-%d")[1:], cmap=sns.color_palette("BuGn_r"))
```

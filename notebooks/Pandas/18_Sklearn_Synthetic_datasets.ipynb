# 18 Sklearn Synthetic datasets

```python
import matplotlib.pyplot as plt
from sklearn import datasets
import numpy as np
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

## Clusters

```python
X, y = datasets.make_blobs(random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X, y = datasets.make_blobs(centers=4, cluster_std=2.0, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X, y = datasets.make_blobs(n_samples=200, centers=5, cluster_std=.75, random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

## Regression

```python
X, y = datasets.make_regression(n_samples=100, n_informative=1, n_features=1, random_state=0)
```

```python
plt.scatter(X, y)
```

```python
X, y, c = datasets.make_regression(n_samples=100, n_informative=1, n_features=1, coef=True, random_state=0)
```

```python
c
```

```python
X_line = np.linspace(-2, 2, 100)
y_line = X_line * c
```

```python
plt.scatter(X, y)
plt.scatter(X_line, y_line)
```

```python
X, y, c = datasets.make_regression(n_samples=100, n_informative=1, n_features=1, noise=20, coef=True, random_state=0)
```

```python
X_line = np.linspace(-2, 2, 100)
y_line = X_line * c
plt.scatter(X, y)
plt.scatter(X_line, y_line)
```

```python
bias = 100
```

```python
X, y, c = datasets.make_regression(n_samples=100, n_informative=1, n_features=1,
                                   bias=bias, noise=20, coef=True, random_state=0)
```

```python
X_line = np.linspace(-2, 2, 100)
y_line = X_line * c + bias
plt.scatter(X, y)
plt.scatter(X_line, y_line)
```

## Classification

```python
X, y = datasets.make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1,
                                    random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X, y = datasets.make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=1, n_clusters_per_class=1,
                                    random_state=0)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X, y = datasets.make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=5,
                                    n_clusters_per_class=1,
                                    random_state=0)
```

```python
X.shape
```

```python
np.unique(y)
```

### Visualization with PCA

```python
pca = PCA(n_components=2, random_state=0)
```

```python
X_reduced = pca.fit_transform(X)
```

```python
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y)
```

### Visualization wit T-SNE

```python
from sklearn.manifold import TSNE
```

```python
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
```

```python
X_tsne = tsne.fit_transform(X)
```

```python
X_tsne.shape
```

```python
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)
```
